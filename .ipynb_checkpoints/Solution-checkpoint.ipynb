# Model 

```python
X = iris.drop(['Species','label'],axis=1)
y = iris['label']
std_x = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(std_x,y,test_size=0.2,random_state=SEED)
```

```python
clf = LogisticRegression(multi_class='auto',random_state=SEED).fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python
accuracy_score(y_test,clf.predict(X_test))
```
